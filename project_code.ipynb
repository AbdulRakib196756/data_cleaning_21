```python
# check the loan and free value 
for index ,row in fifa.iterrows(): 
    if 'On Loan' in row['Contract'] or 'Free' in row['Contract']:
        print(row['Contract'])
```

```python
def extract_contract_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        start_date=np.nan
        end_date=np.nan
        contract_length=0
    else:
        start_date,end_date=contract.split(' ~ ')
        start_year= int(start_date[:4])
        end_year=int(end_date[:4])
        contract_length=end_year-start_year
    return start_date,end_date,contract_length        

# apply fun to colamn and add new column 

new_cols = ['Contract Start', 'Contract End', 'Contract length(years)']

new_data = fifa['Contract'].apply(lambda x: pd.Series(extract_contract_info(x)))

for i in range(len(new_cols)):
    fifa.insert(loc=fifa.columns.get_loc('Contract') + 1 + i, column=new_cols[i], value=new_data[i])    
```

```python
fifa[['Contract','Contract Start','Contract End','Contract length(years)']].sample(5)
```

```python
# contract catagories 
def catagories_contract_status(contract):
    if contract=='Free':
        return "Free"
    elif 'On Loan' in contract:
        return "On Loan"
    else:
        return 'Contract'
    
fifa.insert(fifa.columns.get_loc('Contract length(years)')+1,'Contract Status',fifa['Contract'].apply(catagories_contract_status))    
```

```python
# convert fit and inch into cm 
def convert_height(height):
    if "cm" in height:
        return int(height.strip("cm"))
    else:
        feet, inch = height.split("'")
        total_inch = int(feet) * 12 + int(inch.strip('"'))
        return round(total_inch * 2.54)

# apply func 
fifa['Height'] = fifa['Height'].apply(convert_height)
    


```

```python
fifa['Height'].unique()
```

```python
fifa=fifa.rename(columns={'Height':'Height(cm)'})
fifa
```

```python
def convert_weight (weight):
    if "kg" in weight:
        return int(weight.strip("kg"))
    else:
        lb=int(weight.strip("lbs"))
        return round(lb/2.20)
    
fifa['Weight']=fifa['Weight'].apply(convert_weight) 
fifa['Weight'].unique()   
         
```

```python
# rename wight 
fifa=fifa.rename(columns={'Weight':'Weight(kg)'})
fifa
```

```python
def convert_hits(hits):
    if isinstance(hits, str) and "K" in hits:
        value = float(hits.strip("K"))
        return value * 1000
    else:
        return float(hits) if isinstance(hits, str) else hits

# Assuming you have a DataFrame called 'fifa' with a 'Hits' column
fifa['Hits'] = fifa['Hits'].apply(convert_hits)
```
